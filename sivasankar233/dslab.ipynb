```python
from sklearn.tree import DecisionTreeClassifier
```

```python
x = ds.drop(columns=('diagnosis'))
```

```python
y = ds['diagnosis']
```

```python
model = DecisionTreeClassifier(criterion='entropy')
```

```python
from sklearn import tree
```

```python
tree.plot_tree(model,class_names=y,rounded=True)
```

```python
import matplotlib.pyplot as plt
```

```python
barWidth = 0.25
fig = plt.subplots(figsize =(12, 8))
```

```python
men = [22, 30, 35, 35, 26]
women = [25, 32, 30, 35, 29]
```

```python
br1 = np.arange(len(men))
br2 = [x + barWidth for x in br1]
```

```python
plt.bar(br1, men, color ='b', width = barWidth,
        edgecolor ='grey', label ='men')
plt.bar(br2, women, color ='y', width = barWidth,
        edgecolor ='grey', label ='women')
plt.xlabel('Scores', fontweight ='bold', fontsize = 15)
plt.ylabel('person', fontweight ='bold', fontsize = 15)
plt.title('scoresby person')
plt.xticks([r + barWidth for r in range(len(men))],
        ['G1', 'G2', 'G3', 'G4', 'G5'])
plt.show()
```
