```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
Means_Men=[22,30,35,35,26]
Means_Women=[25,32,30,35,29]
```

```python
n=5
fig, ax = plt.subplots()
index = np.arange(n)
bar_width = 0.3
opacity = 0.7

Men = plt.bar(index, Means_Men, bar_width,
alpha=opacity,
color='blue',
label='Men')

Women = plt.bar(index + bar_width, Means_Women, bar_width,
alpha=opacity,
color='orange',
label='Women')

plt.xlabel('Person')
plt.ylabel('Scores')
plt.title('Scores by person')
plt.xticks(index + bar_width, ('G1', 'G2', 'G3', 'G4', 'G5'))
plt.legend()
plt.show()
```
