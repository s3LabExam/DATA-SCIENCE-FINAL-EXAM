```python
import pandas as pd
import numpy as np
```

```python
data=pd.read_csv('data.csv')
```

```python
data
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc=StandardScaler()
```

```python
new_data=sc.fit_transform(data.drop('diagnosis',axis=1))
```

```python
from sklearn.model_selection import train_test_split
```

```python
x=new_data
y=data['diagnosis']
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
classifier1=DecisionTreeClassifier(criterion="entropy")
```

```python
classifier1.fit(x,y)
```

```python
from sklearn import tree
```

```python
tree.plot_tree(classifier1,filled=True,class_names=y,rounded=True)
```

```python
pred = classifier1.predict([[0,0,1,0,0]])
```
