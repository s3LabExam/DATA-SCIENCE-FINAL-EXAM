```python
mat1=[]
mat2=[]
```

```python
def read(m):
  l=[]
  for i in range(n):
    for j in range(n):
      l.append(int(input()))
    m.append(l[-n:])
  
  
```

```python
def print_mat(m):
  for i in m:
    for j in i:
      print(j,end="   ")
    print("\n")
```

```python
read(mat1)
```

```python
print_mat(mat1)
  
```
