```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
num = 5
men= (22, 30, 33, 30, 26)
women= (25, 32, 30, 35, 29)
```

```python
fig, ax = plt.subplots()
ind = np.arange(num)
bar_width = 0.2
opacity = 0.9
```

```python
ax1 = plt.bar(ind, men, bar_width,
alpha=opacity,
color='r',
label='Men')

ax2 = plt.bar(ind + bar_width, women, bar_width,
alpha=opacity,
color='y',
label='Women')

plt.xlabel('Person')
plt.ylabel('Scores')
plt.title('Scores by person')

plt.xticks(ind + bar_width, ('G1', 'G2', 'G3', 'G4', 'G5'))
plt.legend(loc='best')

plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
df =  pd.read_csv('medal.csv')
country_data = df["country"]
medal_data = df["gold_medal"]
colors = ["blue", "green", "cyan", "red", "yellow"]
explode = (0.1, 0, 0, 0, 0)  
plt.pie(medal_data, labels=country_data, explode=explode, colors=colors,
autopct='%1.1f%%', shadow=True, startangle=100)
plt.title("Gold medal achievements of five most successful countries in 2016 Summer Olympics")
plt.show()
```
